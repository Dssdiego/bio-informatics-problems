```python
rev = list(reversedString)
result = list()

for i in range(len(rev)):
    c = rev[i]
    if(c == "A"):
        c = "T"
    elif (c == "T"):
        c = "A"
    elif (c == "C"):
        c = "G"
    elif (c == "G"):
        c = "C"
    result.insert(i, c)
    i += 1
```
